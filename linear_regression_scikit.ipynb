```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
```

```python
df = load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
x = df[['bmi']]
y = df['target']
```

```python
reg = LinearRegression().fit(x, y)
```

```python
def create_plot(x, y, x_line=None, y_line=None):
    # Create the scatter plot
    fig = px.scatter(x=x, y=y, title="Body Mass Index (BMI) vs. Diabetes Progression Target")

    fig.update_layout(
        xaxis_title="Body Mass Index (BMI)",
        yaxis_title="Diabetes Progression Target"
    )

    # Add the best fit line if the values are provided
    if x_line is None or y_line is None:
        return fig
    
    fig.add_trace(
        go.Scatter(
            x=x_line,
            y=y_line,
            mode='lines',
            name='Best Fit Line',
            line=dict(color='red', width=2)
        )
    )
    return fig
```

```python
x_line = np.linspace(min(df['bmi']), max(df['bmi']), 10)
y_line = reg.coef_ * x_line + reg.intercept_
fig = create_plot(df['bmi'], y, x_line, y_line)
fig.show()
```
